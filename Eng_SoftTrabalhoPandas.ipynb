```python
#Adicionar coluna total
df['Total'] = df['Ataque'] + df['Ataque'] + df['Def Esp'] + df['Defensa'] + df['Velocidade'] + df['Vida']
df
```

Remover coluna

```python
#Remover coluna
df = df.drop(columns=['Total'])
df
```

Forma alternativa de adicionar coluna

```python
#não esquecer que sempre vai imprimir um a menos [:, 4:10] no caso áte oa coluna 9
from pandas.core.frame import Axis
df['Total'] = df.iloc[:, 4:10] .sum(axis=1)#todas as linhas ,coluna 4 até 10 [:, 4:10] axis=1 eixo das colunas
df
```
